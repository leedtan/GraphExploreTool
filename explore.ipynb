```python
import json
import pandas as pd
import numpy as np
with open('data') as handle:
    dictdump = json.loads(handle.read())
```

```python
df = pd.DataFrame(dictdump)
```

```python
df.head()
```

```python
source_usrs = set(df['source'].unique())
target_usrs = set(df['target'].unique())
all_uni_usrs = source_usrs.union(target_usrs)
def build_forward_backward_hash(unique_values):
    num_unique = len(unique_values)
    fmap = dict(zip(unique_values, range(num_unique)))
    bmap = dict(zip(range(num_unique), unique_values))
    return fmap, bmap
```

```python
usr2idx, idx2usr = build_forward_backward_hash(all_uni_usrs)
```

```python
all_nodes = {}
```

```python
def ensure_nodes(nodes):
    for node in nodes:
        if not node in all_nodes:
            all_nodes[node] = {'connections': set(), 's': set(), 't': set()}
    return
```

```python
def update_node(active, relative, dp, connection_type):
    all_nodes[active][connection_type].add(dp)
    all_nodes[active]['connections'].add(relative)
```

```python
def process_connection(dp, source, target):
    ensure_nodes([source, target])
    update_node(source, target, dp, 's')
    update_node(target, source, dp, 't')
```

```python
for idx in range(df.shape[0]):
    if idx % 10000 == 0:
        print('%d of %d, progress: %.3f percent ' % (idx, df.shape[0], idx / df.shape[0]) )

        #print(idx, 'of', df.shape[0], ', progress: ', idx / df.shape[0],'%' )
    process_connection(*df.loc[idx,:])
```
